```
import torch
from diffusers import StableDiffusionXLPipeline
from IPython.display import display, Image as IPImage
import os

print('Chargement SDXL...')
pipe = StableDiffusionXLPipeline.from_pretrained(
    'stabilityai/stable-diffusion-xl-base-1.0',
    torch_dtype=torch.float16,
    variant='fp16',
    use_safetensors=True
).to('cuda')
print('‚úÖ Mod√®le charg√©!')
```

```
os.makedirs('/content/chat-violet-refs', exist_ok=True)

# Style ultra d√©taill√© bas√© sur l'image de r√©f√©rence
style = '2D western cartoon style, adult swim aesthetic, thick bold black outlines, matte flat solid colors with no gradients, simple shading, no anime, no cute, no kawaii, no 3D, hand-drawn look, like Aqua Teen Hunger Force or Superjail art style'

# Personnage ultra d√©taill√©
character = 'a very fat obese round-bodied deep indigo-purple cat with an extremely grumpy annoyed unimpressed expression, heavy droopy half-closed eyelids with visible pink-salmon eyeshadow around the eyes, small dark purple triangular nose, tiny frowning mouth, pointy triangular ears with pink inner ear, short stubby thick limbs, blob-like round body shape, no neck just a big round torso, the cat looks perpetually tired bored and judgmental'

prompts = [
    f'{character}, sitting on a grey couch holding a TV remote looking bored, {style}',
    f'{character}, standing upright with arms crossed looking angry and judging, full body view, {style}',
    f'{character}, sleeping in a tight ball with annoyed expression even while sleeping, {style}',
    f'{character}, aggressively eating from a food bowl, crumbs everywhere, {style}',
    f'{character}, front facing portrait staring directly at viewer with maximum judgment, {style}',
    f'{character}, side profile view showing round belly and stubby legs, {style}',
    f'{character}, squeezed into a tiny cardboard box, fat spilling over edges, annoyed, {style}',
    f'{character}, wearing a tiny golden crown sitting on a toilet like a throne, unimpressed, {style}',
    f'{character}, holding a dead fish by the tail dangling it with one paw, smug expression, {style}',
    f'{character}, mid-yawn showing sharp teeth and pink tongue, eyes squinted, {style}',
    f'{character}, pushing a coffee mug off a table with one paw while staring at viewer, {style}',
    f'{character}, wearing an oversized white lab coat and tiny safety goggles, holding a bubbling flask, mad scientist, {style}',
    f'{character}, floating in dark space surrounded by glowing particles and equations, existential, {style}',
    f'{character}, simultaneously inside and outside a cardboard box, ghostly transparent duplicate, quantum superposition concept, {style}',
    f'{character}, extreme closeup face filling the frame, every detail of the grumpy expression visible, {style}',
    f'{character}, walking away from viewer looking back over shoulder with contempt, {style}',
]

negative = 'anime, manga, kawaii, cute, chibi, 3D render, realistic, photographic, soft shading, gradient colors, pastel, smooth, pixar, disney, blurry, deformed, watercolor, painterly'

for i, prompt in enumerate(prompts):
    print(f'Image {i+1}/{len(prompts)}')
    image = pipe(prompt=prompt, negative_prompt=negative, num_inference_steps=30, guidance_scale=7.5, width=1024, height=1024).images[0]
    path = f'/content/chat-violet-refs/chat_violet_{i:02d}.png'
    image.save(path)
    display(IPImage(filename=path, width=300))

print(f'\n‚úÖ {len(prompts)} images sauvegard√©es!')
```
