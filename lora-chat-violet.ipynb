```
import torch
from diffusers import StableDiffusionXLPipeline
from IPython.display import display, Image as IPImage
import os

print('Chargement SDXL...')
pipe = StableDiffusionXLPipeline.from_pretrained(
    'stabilityai/stable-diffusion-xl-base-1.0',
    torch_dtype=torch.float16,
    variant='fp16',
    use_safetensors=True
).to('cuda')
print('‚úÖ Mod√®le charg√©!')
```

```
os.makedirs('/content/chat-violet-refs', exist_ok=True)

base_style = 'cartoon style, bold outlines, flat colors, animated series aesthetic, thick black outlines'
character = 'a grumpy chubby purple cat with half-closed sleepy eyes, pink inner ears, small dark purple nose, round body'

prompts = [
    f'{character} sitting on a grey couch looking annoyed, {base_style}',
    f'{character} standing with arms crossed looking angry, full body, {base_style}',
    f'{character} sleeping curled up in a ball, {base_style}',
    f'{character} eating from a food bowl greedily, {base_style}',
    f'{character} looking at camera with judgmental expression, front view portrait, {base_style}',
    f'{character} side profile view, {base_style}',
    f'{character} sitting in a cardboard box that is too small, {base_style}',
    f'{character} wearing a tiny golden crown looking unimpressed, {base_style}',
    f'{character} holding a fish in its paw looking smug, {base_style}',
    f'{character} yawning wide showing teeth, {base_style}',
    f'{character} pushing things off a table with its paw, {base_style}',
    f'{character} wearing a lab coat and safety goggles, scientist, {base_style}',
    f'{character} floating in space surrounded by glowing quantum particles, {base_style}',
    f'{character} both inside and outside a box, quantum superposition concept, {base_style}',
    f'{character} closeup face portrait detailed expression, {base_style}',
    f'{character} walking away looking back over shoulder, {base_style}',
]

negative = 'realistic, photographic, 3d render, blurry, low quality, deformed, ugly, bad anatomy'

for i, prompt in enumerate(prompts):
    print(f'Image {i+1}/{len(prompts)}')
    image = pipe(prompt=prompt, negative_prompt=negative, num_inference_steps=30, guidance_scale=7.5, width=1024, height=1024).images[0]
    path = f'/content/chat-violet-refs/chat_violet_{i:02d}.png'
    image.save(path)
    display(IPImage(filename=path, width=300))

print(f'\n‚úÖ {len(prompts)} images sauvegard√©es!')
```
